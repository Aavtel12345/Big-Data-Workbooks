```python
import pandas as pd
from sklearn import metrics
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
```

```python
data = pd.read_csv("H:/share/Science Policy Portfolio/Census/nihTopics.csv", low_memory=False)
#data.to_csv("H:/share/Science Policy Portfolio/Census/nihSmall.csv")
```

```python
data.describe()
```

```python
data_valid = data[data["ORG_DEPT"] != "NONE"]
data_valid = data_valid[pd.notnull(data_valid['ORG_DEPT'])]
#data_valid = data_valid[["CFDA_CODE","STUDY_SECTION","TOTAL_COST", "ORG_NAME","ORG_DEPT", "ARRA_FUNDED", "topic_id"]]
#data_valid = data_valid.dropna()
```

```python
data_valid.describe()
```

```python
data_x = data_valid[["CFDA_CODE","STUDY_SECTION","TOTAL_COST", "ORG_NAME", "ARRA_FUNDED", "topic_id"]]
data_y = data_valid["ORG_DEPT"]
data_y.describe()
```

```python
data_x1 = pd.get_dummies(data_x)
data_train_x = data_x1[:25000]
data_train_y = data_y[:25000]
data_test_x = data_x1[25001:]
data_test_y = data_y[25001:]
from sklearn.metrics import accuracy_score
```

```python
#Decision tree
model = DecisionTreeClassifier()
model.fit(data_x1, data_y)
print(model)
expected = data_test_y
predicted = model.predict(data_test_x)
# summarize the fit of the model
print(metrics.classification_report(expected, predicted))
print(metrics.confusion_matrix(expected, predicted))
plt.matshow(metrics.confusion_matrix(expected, predicted))
plt.title('Confusion matrix')
plt.colorbar()
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.show()
```

```python
#Multinomial logit
model = LogisticRegression()
model.fit(data_x1, data_y)
print(model)
expected = data_test_y
predicted = model.predict(data_test_x)
# summarize the fit of the model
#print(metrics.classification_report(expected, predicted))
#print(metrics.confusion_matrix(expected, predicted))
#plt.matshow(metrics.confusion_matrix(expected, predicted))
#plt.title('Confusion matrix')
#plt.colorbar()
#plt.ylabel('True label')
#plt.xlabel('Predicted label')
#plt.show()
accuracy_score(expected, predicted)
```

```python
#Random Forest
model = RandomForestClassifier()
model.fit(data_x1, data_y)
print(model)
expected = data_test_y
predicted = model.predict(data_test_x)
# summarize the fit of the model
print(metrics.classification_report(expected, predicted))
print(metrics.confusion_matrix(expected, predicted))
plt.matshow(metrics.confusion_matrix(expected, predicted))
plt.title('Confusion matrix')
plt.colorbar()
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.show()
```
